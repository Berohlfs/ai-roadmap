````python
import matplotlib.pyplot as plt
import numpy as np

# Simple matplotlib graph
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, 'b-', linewidth=2, label='sin(x)')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Simple Matplotlib Graph')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
````
